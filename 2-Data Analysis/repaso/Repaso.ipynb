## Parentesis vs Corchetes
1. **Corchetes** (seleccionar)
    - Listas, tuplas: indexing o slicing
    - Numpy: index y slicing
    - Dataframe: sobre DF o con loc e iloc
2. **Parentesis**
    - Declaraciones: funciones, clases (metodos, __init__....)
    - Llamadas a funciones
    - Llamadas a metodos de la clase

## Apply

```python
df.columns
df['New col'] = df['Sales'] * df['Quantity']
df.head()
```

```python
df['Sales a tope'] = df['Sales'].apply(lambda x: x*1000)
```

```python
def mifun(x):
    return x*1000

df['Sales'].apply(mifun)
```

```python
def mifun(x):
    if x == 'Kentucky':
        return 1
    elif x == 'California':
        return 2
    else:
        return 0

df['State'].apply(mifun).unique()
```

```python
def mifun(x, y):
    return x + ' - ' + y

df.apply(lambda x: mifun(x['City'], x['State']), axis=1)
```

```python
def mifun(x):
    return pd.Series((x*10, x*1000))

df['Sales'].apply(mifun)
```
