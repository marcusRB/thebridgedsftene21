```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', 
                      sep='|', index_col='user_id')
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.groupby(['occupation', 'gender']).count().reset_index()[['occupation', 'gender', 'age']]
```

```python
users2 = users.groupby(['occupation', 'gender']).count()

users2 = users2.reset_index()[['occupation', 'gender', 'age']]
users2.rename(columns={'age': 'Conteo Occ'}, inplace=True)
users2
```

```python
def get_ratio(x):
    x['ratio'] = x['Conteo Occ']/x['Conteo Occ'].sum()
    
users2.groupby('occupation').apply(get_ratio)
```

```python
# Good one and easy
def norm_by_data2(x):
    x['ratio'] = x['age']*100/x['age'].sum()
    return x

users2 = users.groupby(['occupation', 'gender']).count()
users2.reset_index().groupby(['occupation']).apply(norm_by_data2)[['occupation', 'gender', 'ratio']]
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg(['min', 'max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation', 'gender']).mean().reset_index()
```
