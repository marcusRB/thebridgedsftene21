## Instalation & Configuration steps:

1. Install on Windows (third point with Mac/Linux):

    - https://git-scm.com/download/win

    - After install by default, lauch git bash

    - Looking for "Terminal" on Mac(On Mac/Linux only run "git" on command line and install from there)

2. Create a github account:

    - https://github.com/

3. Clone (http) the Repository Link:

    - Create a "DataScience" folder on your OS (no long paths)

    - With command line (bash on Windows), move the pointer to the folder created and write there: 
    
            git clone repository_url(ask teacher for the repository url)
<br>
    - With this, you can do "pull" (inside command line (bash) inside project's folder) to upload the data with the last commits.

4. Create your own repository on github:
    
    - Choose a name, write a description, make it public and initialize a Readme

5. Create your own SSH key:

    - Create your own ssh name and mail:
    
            git config --global user.email "MY_NAME@example.com"
            git config --global user.name "NAME"

    - Write in the command line (bash) with your github mail:

            ssh-keygen -t rsa -C "your_email@example.com"

        Choose by default the directory (remember it) when you are creating the ssh keygen. Choose your own password. 

    - Open the directory where you have created the ssh key.

    - Open, see and copy your entire public ssh key generated "id_rsa.pub"
    
6. Add your public ssh key in your github:
    - With the public ssh key copied, go to your github and go to the "Settings" clicking in your profile on the top-right corner and selecting "Settings". 

    - There, go to the "SSH and GPG Keys" section on the left.

    - Click on the green button "New SSH key".

    - Add a title and paste your public SSH key.

    - Finish with "Add SSH key".
    
7. Cloning your own repository:

    - Go to your repository page.

    - Click on "Clone or download" your repository, choose SSH option and copy the link.

    - Make a new folder for your repository in your OS. Go to that folder with the command line (bash on windows) and write:

            git clone "Paste_here_your_repository_link"

8. Checking all is OK:

    - Go to your folder repository on your OS with the command line (bash) and write: 

            git status

        If there are no errors, then it is OK to start to work!
